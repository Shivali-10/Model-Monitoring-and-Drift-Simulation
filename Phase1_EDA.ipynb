```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

print("All libraries imported successfully!")
```

```python
df = pd.read_csv('creditcard.csv')
print("Dataset loaded successfully!")
print("Shape:", df.shape)
```

```python
print(df.head())
```

```python
print(df.info())
```

```python
print(df['Class'].value_counts())
```

```python
sns.countplot(x='Class', data=df)
plt.title('Fraud vs Non-Fraud')
plt.xticks([0,1], ['Non-Fraud', 'Fraud'])
plt.show()
```

```python
print("Average amount - Non-Fraud:", df[df['Class']==0]['Amount'].mean().round(2))
print("Average amount - Fraud:", df[df['Class']==1]['Amount'].mean().round(2))
print("Max amount - Non-Fraud:", df[df['Class']==0]['Amount'].max())
print("Max amount - Fraud:", df[df['Class']==1]['Amount'].max())
```

```python
plt.figure(figsize=(10,4))
plt.hist(df[df['Class']==0]['Amount'], bins=50, alpha=0.6, color='blue', label='Non-Fraud')
plt.hist(df[df['Class']==1]['Amount'], bins=50, alpha=0.6, color='red', label='Fraud')
plt.title('Transaction Amount - Fraud vs Non-Fraud')
plt.xlabel('Amount')
plt.ylabel('Count')
plt.legend()
plt.show()
```
